```python
#Utilizando pandas
import pandas as pd


# Función para leer un archivo CSV desde una URL usando Pandas
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en un DataFrame de Pandas.
    """
    try:
        df = pd.read_csv(url)
        print(f"Datos cargados exitosamente desde {url}")
        return df
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return pd.DataFrame()  # Retorna un DataFrame vacío en caso de error

# Función para calcular el ingreso total de una tienda
def calcular_ingreso_total(df, columna_precio="Precio"):
    """
    Calcula el ingreso total sumando los valores de la columna de precios.

    Parámetros:
        df (pd.DataFrame): El DataFrame que contiene los datos de la tienda.
        columna_precio (str): El nombre de la columna que contiene los precios.

    Retorna:
        float: El ingreso total calculado.
    """
    if columna_precio not in df.columns:
        print(f"La columna '{columna_precio}' no existe en el DataFrame.")
        return None

    ingreso_total = df[columna_precio].sum()
    return ingreso_total

#Función para calcular la sumatoria de la facturación de todas las tiendas
def calcular_sumatoria_facturacion(ingresos_totales):
    """
    Calcula la sumatoria de los ingresos totales de todas las tiendas.

    Parámetros:
        ingresos_totales (dict): Un diccionario donde las claves son los nombres de las tiendas
                                 y los valores son los ingresos totales de cada tienda.

    Retorna:
        float: La sumatoria de los ingresos totales de todas las tiendas.
    """
    if not ingresos_totales:
        print("No hay datos de ingresos disponibles.")
        return None

    sumatoria = sum(ingresos_totales.values())
    return sumatoria

# URLs de los archivos CSV
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv "
]

# Calcular el ingreso total para cada tienda
ingresos_totales = {}
for i, url in enumerate(urls, start=1):
    # Leer el CSV
    tienda_df = leer_csv(url)
    if tienda_df is not None:
        # Calcular el ingreso total
        ingreso_total = calcular_ingreso_total(tienda_df, columna_precio="Precio")
        if ingreso_total is not None:
            ingresos_totales[f"Tienda {i}"] = ingreso_total

# Mostrar los ingresos totales de cada tienda
print("Ingresos Totales por Tienda:")
for tienda, ingreso in ingresos_totales.items():
    print(f"{tienda}: ${ingreso:.2f}")

# Calcular la sumatoria de la facturación de todas las tiendas
sumatoria_facturacion = calcular_sumatoria_facturacion(ingresos_totales)
if sumatoria_facturacion is not None:
    print(f"\nFacturación Total de Todas las Tiendas: ${sumatoria_facturacion:.2f}")
```

```python
import csv
from urllib.request import urlopen

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url.strip()) as respuesta:  # Eliminamos espacios adicionales en la URL
            # Decodificar los datos
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para calcular el ingreso total de una tienda
def calcular_ingreso_total(data, columna_precio='Precio'):
    """
    Calcula el ingreso total sumando los valores de la columna 'Precio' de los datos.

    Args:
    - data: Lista de diccionarios que contiene los datos de la tienda.
    - columna_precio: Nombre de la columna que contiene los precios (por defecto es 'Precio').

    Returns:
    - Ingreso total como un número flotante.
    """
    ingreso_total = 0
    for row in data:
        try:
            # Convertir el precio a float y sumarlo al ingreso total
            precio = float(row[columna_precio])
            ingreso_total += precio
        except KeyError:
            print(f"Error: La columna '{columna_precio}' no existe en los datos.")
            return None
        except ValueError:
            print(f"Error: El valor '{row[columna_precio]}' no es un número válido.")
            return None
    return ingreso_total

# Función para calcular la sumatoria de ingresos de todas las tiendas
def calcular_sumatoria_ingresos(ingresos_list):
    """
    Calcula la sumatoria de los ingresos totales de todas las tiendas.

    Args:
    - ingresos_list: Lista de ingresos totales de cada tienda.

    Returns:
    - Sumatoria de los ingresos totales como un número flotante.
    """
    if not ingresos_list:
        print("La lista de ingresos está vacía.")
        return 0.0

    sumatoria = sum(ingresos_list)
    return sumatoria

# URLs de las tiendas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv    ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv    ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv    ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv    "
]

# Calcular el ingreso total para cada tienda
ingresos_totales = []
for i, url in enumerate(urls, start=1):
    # Leer los datos desde la URL
    data = leer_csv(url)
    if data:
        # Calcular el ingreso total
        ingreso_total = calcular_ingreso_total(data)
        if ingreso_total is not None:
            ingresos_totales.append(ingreso_total)
            print(f"Ingreso total de la Tienda {i}: ${ingreso_total:.2f}")
        else:
            print(f"No se pudo calcular el ingreso total para la Tienda {i}.")
    else:
        print(f"No se pudieron leer los datos de la Tienda {i}.")

# Resultado final
print("\nResumen de ingresos totales:")
for i, ingreso in enumerate(ingresos_totales, start=1):
    print(f"Tienda {i}: ${ingreso:.2f}")

# Calcular la sumatoria de ingresos de todas las tiendas
sumatoria_ingresos = calcular_sumatoria_ingresos(ingresos_totales)
print(f"\nSumatoria de Ingresos de Todas las Tiendas: ${sumatoria_ingresos:.2f}")
```

```python
import pandas as pd

# Función para leer un archivo CSV desde una URL usando Pandas
def leer_csv_pandas(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en un DataFrame de Pandas.
    """
    try:
        df = pd.read_csv(url)
        print(f"Datos cargados exitosamente desde {url}")
        return df
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return pd.DataFrame()  # Retorna un DataFrame vacío en caso de error

# Función para contar productos vendidos por categoría
def contar_productos_por_categoria(df):
    """
    Recibe un DataFrame y cuenta los productos vendidos por categoría.
    Retorna un diccionario con las categorías como claves y los conteos como valores.
    """
    if "Categoría del Producto" not in df.columns:
        print("La columna 'Categoría del Producto' no está presente en los datos.")
        return {}
    conteo_categorias = df["Categoría del Producto"].value_counts().to_dict()
     # Ordenar por cantidad de productos vendidos (descendente)
    conteo_categorias = dict(sorted(conteo_categorias.items(), key=lambda x: x[1], reverse=True))
    return conteo_categorias

# Función para calcular el total de ventas en dinero por categorías
def calcular_ventas_por_categorias(df):
    """
    Calcula el total de ventas en dinero por categorías.
    Retorna un diccionario con las categorías como claves y los totales de ventas como valores.
    """
    try:
        # Asegurarse de que las columnas necesarias existan
        if "Categoría del Producto" not in df.columns or "Precio" not in df.columns or "Cantidad de cuotas" not in df.columns:
            print("Faltan columnas necesarias para calcular las ventas por categorías.")
            return {}

        # Calcular el monto de venta por fila
        df["Monto Venta"] = df["Precio"] * df["Cantidad de cuotas"]

        # Agrupar por categoría y sumar los montos de venta
        ventas_por_categorias = df.groupby("Categoría del Producto")["Monto Venta"].sum().to_dict()
        # Ordenar por monto total de ventas (descendente)
        ventas_por_categorias = dict(sorted(ventas_por_categorias.items(), key=lambda x: x[1], reverse=True))
        return ventas_por_categorias
    except Exception as e:
        print(f"Error al calcular las ventas por categorías: {e}")
        return {}

# Función para calcular el total de ventas en dinero
def calcular_total_ventas(df):
    """
    Calcula el total de ventas en dinero sumando el precio de cada producto multiplicado por la cantidad de cuotas.
    Retorna el total de ventas.
    """
    try:
        if "Precio" not in df.columns or "Cantidad de cuotas" not in df.columns:
            print("Faltan columnas necesarias para calcular el total de ventas.")
            return 0
        total_ventas = (df["Precio"] * df["Cantidad de cuotas"]).sum()
        return total_ventas
    except Exception as e:
        print(f"Error al calcular el total de ventas: {e}")
        return 0

# Función para mostrar los resultados
def mostrar_resultados(resultados, totales_ventas, ventas_por_categorias):
    """
    Muestra los conteos de productos por categoría, el total de ventas en dinero,
    y el total de ventas por categorías.
    """
    for tienda, conteo in resultados.items():
        print(f"\n{tienda}:")
        if not conteo:  # Si el conteo está vacío
            print("  - No se encontraron datos de ventas.")
        else:
            for categoria, cantidad in conteo.items():
                print(f"  - {categoria}: {cantidad} productos vendidos")

        # Mostrar el total de ventas por categorías
        print("  - Ventas por categorías:")
        for categoria, total_venta in ventas_por_categorias[tienda].items():
            print(f"    - {categoria}: ${total_venta:,.2f}")

        # Mostrar el total de ventas general
        print(f"  - Total de ventas: ${totales_ventas[tienda]:,.2f}")

# Función principal
def main():
    # Opción 1: Usar URLs (comentar si se usan archivos locales)
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv    "
    ]

    # Procesar los datos de todas las tiendas
    resultados = {}
    totales_ventas = {}
    ventas_por_categorias = {}
    for i, url in enumerate(urls, start=1):  # Cambiar `urls` por `rutas_locales` si se usan archivos locales
        nombre_tienda = f"Tienda {i}"

        # Leer los datos de la tienda
        df_tienda = leer_csv_pandas(url.strip())  # Eliminar espacios adicionales en la URL

        if df_tienda.empty:  # Si el DataFrame está vacío, saltar esta tienda
            print(f"No se pudo cargar datos para {nombre_tienda}.")
            continue

        # Contar productos por categoría
        conteo_categorias = contar_productos_por_categoria(df_tienda)
        resultados[nombre_tienda] = conteo_categorias

        # Calcular el total de ventas en dinero
        total_venta = calcular_total_ventas(df_tienda)
        totales_ventas[nombre_tienda] = total_venta

        # Calcular el total de ventas por categorías
        ventas_por_categoria = calcular_ventas_por_categorias(df_tienda)
        ventas_por_categorias[nombre_tienda] = ventas_por_categoria

    # Mostrar los resultados
    mostrar_resultados(resultados, totales_ventas, ventas_por_categorias)

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import csv
from urllib.request import urlopen

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url.strip()) as respuesta:  # Eliminamos espacios adicionales en la URL
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para contar productos vendidos por categoría
def contar_productos_por_categoria(datos):
    """
    Recibe una lista de datos (diccionarios) y cuenta los productos vendidos por categoría.
    Retorna un diccionario con las categorías como claves y los conteos como valores.
    """
    conteo_categorias = {}
    for fila in datos:
        if "Categoría del Producto" not in fila or "Cantidad de cuotas" not in fila:
            print("Las columnas 'Categoría del Producto' o 'Cantidad de cuotas' no están presentes en los datos.")
            return {}
        categoria = fila["Categoría del Producto"]
        cantidad = int(fila["Cantidad de cuotas"])  # Convertimos a entero
        if categoria not in conteo_categorias:
            conteo_categorias[categoria] = 0
        conteo_categorias[categoria] += cantidad
    return conteo_categorias

# Función para calcular el total de ventas en dinero por categorías
def calcular_ventas_por_categorias(datos):
    """
    Calcula el total de ventas en dinero por categorías.
    Retorna un diccionario con las categorías como claves y los totales de ventas como valores.
    """
    ventas_por_categorias = {}
    errores = 0
    for fila in datos:
        try:
            categoria = fila["Categoría del Producto"]
            precio = float(fila["Precio"])
            cantidad = int(fila["Cantidad de cuotas"])
            monto_venta = precio * cantidad
            if categoria not in ventas_por_categorias:
                ventas_por_categorias[categoria] = 0
            ventas_por_categorias[categoria] += monto_venta
        except (ValueError, KeyError):
            errores += 1
            print("Error al procesar el precio o la categoría de un producto. Se omitirá.")
    if errores > 0:
        print(f"Se omitieron {errores} filas debido a errores.")
    return ventas_por_categorias

# Función para mostrar los resultados ordenados
def mostrar_resultados(resultados, totales_ventas, ventas_por_categorias):
    """
    Muestra los conteos de productos por categoría, el total de ventas en dinero,
    y el total de ventas por categorías, ordenados por cantidad vendida y monto total.
    """
    for tienda, conteo in resultados.items():
        print(f"\n{tienda}:")
        if not conteo:  # Si el conteo está vacío
            print("  - No se encontraron datos de ventas.")
            continue

        # Ordenar por cantidad vendida (de mayor a menor)
        conteo_ordenado = sorted(conteo.items(), key=lambda x: x[1], reverse=True)
        print("  - Productos más vendidos por categoría:")
        for categoria, cantidad in conteo_ordenado:
            print(f"    - {categoria}: {cantidad} productos vendidos")

        # Ordenar por monto total de ventas (de mayor a menor)
        ventas_ordenadas = sorted(ventas_por_categorias[tienda].items(), key=lambda x: x[1], reverse=True)
        print("  - Ventas por categorías (monto total):")
        for categoria, total_venta in ventas_ordenadas:
            print(f"    - {categoria}: ${total_venta:,.2f}")

        # Mostrar el total de ventas general
        print(f"  - Total de ventas: ${totales_ventas[tienda]:,.2f}")

# Función principal
def main():
    # URLs de las tiendas
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv  ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv  ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv  ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv  "
    ]

    # Procesar los datos de todas las tiendas
    resultados = {}
    totales_ventas = {}
    ventas_por_categorias = {}
    for i, url in enumerate(urls, start=1):
        nombre_tienda = f"Tienda {i}"
        datos_tienda = leer_csv(url)

        if not datos_tienda:  # Si los datos están vacíos, saltar esta tienda
            print(f"No se pudo cargar datos para {nombre_tienda}.")
            continue

        # Contar productos por categoría
        conteo_categorias = contar_productos_por_categoria(datos_tienda)
        resultados[nombre_tienda] = conteo_categorias

        # Calcular el total de ventas en dinero
        total_venta = sum(float(fila["Precio"]) * int(fila["Cantidad de cuotas"]) for fila in datos_tienda if "Precio" in fila and "Cantidad de cuotas" in fila)
        totales_ventas[nombre_tienda] = total_venta

        # Calcular el total de ventas por categorías
        ventas_por_categoria = calcular_ventas_por_categorias(datos_tienda)
        ventas_por_categorias[nombre_tienda] = ventas_por_categoria

    # Mostrar los resultados
    mostrar_resultados(resultados, totales_ventas, ventas_por_categorias)

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import csv
from urllib.request import urlopen

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url.strip()) as respuesta:  # Eliminamos espacios adicionales en la URL
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para calcular las calificaciones promedio de los clientes
def calcular_calificacion_promedio(datos):
    """
    Calcula la calificación promedio de los clientes para una tienda.
    Retorna la calificación promedio como un número flotante.
    """
    calificaciones = []
    for fila in datos:
        try:
            calificacion = float(fila["Calificación"])
            calificaciones.append(calificacion)
        except (ValueError, KeyError):
            print("Error al procesar la calificación de un cliente. Se omitirá.")

    if not calificaciones:  # Si no hay calificaciones válidas
        print("No se encontraron calificaciones válidas.")
        return None

    promedio = sum(calificaciones) / len(calificaciones)
    return promedio

# Función principal
def main():
    # URLs de las tiendas
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv  ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv  ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv  ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv  "
    ]

    # Procesar los datos de todas las tiendas
    calificaciones_promedio = {}
    for i, url in enumerate(urls, start=1):
        nombre_tienda = f"Tienda {i}"
        datos_tienda = leer_csv(url)

        if not datos_tienda:  # Si los datos están vacíos, saltar esta tienda
            print(f"No se pudo cargar datos para {nombre_tienda}.")
            continue

        # Calcular la calificación promedio
        calificacion_promedio = calcular_calificacion_promedio(datos_tienda)
        calificaciones_promedio[nombre_tienda] = calificacion_promedio

    # Mostrar las calificaciones promedio
    print("\nValoraciones Medias por Tienda:")
    for tienda, calificacion in calificaciones_promedio.items():
        if calificacion is not None:
            print(f"{tienda}: {calificacion:.2f}")
        else:
            print(f"{tienda}: No se encontraron calificaciones válidas.")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import pandas as pd

# Función para leer un archivo CSV desde una URL usando Pandas
def leer_csv_pandas(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en un DataFrame de Pandas.
    """
    try:
        df = pd.read_csv(url.strip())  # Eliminamos espacios adicionales en la URL
        print(f"Datos cargados exitosamente desde {url}")
        return df
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return None

# Función para agregar una columna 'Tienda' a cada DataFrame
def agregar_columna_tienda(df, nombre_tienda):
    """
    Agrega una columna 'Tienda' al DataFrame indicando el nombre de la tienda.

    Parámetros:
        df (pd.DataFrame): El DataFrame original.
        nombre_tienda (str): El nombre de la tienda.

    Retorna:
        pd.DataFrame: El DataFrame con la columna 'Tienda' agregada.
    """
    df['Tienda'] = nombre_tienda
    return df

# Función para calcular las ventas por categoría
def calcular_ventas_por_categoria(datos_combinados):
    """
    Calcula la cantidad de productos vendidos por categoría en cada tienda.

    Parámetros:
        datos_combinados (pd.DataFrame): El DataFrame combinado de todas las tiendas.

    Retorna:
        dict: Un diccionario con las ventas por categoría.
    """
    ventas_por_categoria = (
        datos_combinados.groupby(['Tienda', 'Categoría del Producto'])
        .agg(Ventas=('Cantidad de cuotas', 'sum'))
        .reset_index()
    )
    return ventas_por_categoria

# Función para identificar las categorías más populares
def identificar_categorias_mas_populares(ventas_por_categoria):
    """
    Identifica las categorías más populares de cada tienda.

    Parámetros:
        ventas_por_categoria (pd.DataFrame): El DataFrame con las ventas por categoría.

    Retorna:
        dict: Un diccionario con las categorías más populares.
    """
    categorias_mas_populares = (
        ventas_por_categoria.loc[ventas_por_categoria.groupby('Tienda')['Ventas'].idxmax()]
    )
    return categorias_mas_populares

# Función para identificar los productos más y menos vendidos
def identificar_top_productos(datos_combinados, top_n=3):
    """
    Identifica los productos más vendidos y menos vendidos de cada tienda.

    Parámetros:
        datos_combinados (pd.DataFrame): El DataFrame combinado de todas las tiendas.
        top_n (int): Número de productos a incluir en los resultados.

    Retorna:
        dict: Un diccionario con los productos más y menos vendidos por tienda.
    """
    # Agrupar por Tienda y Producto, y contar el número de ventas
    ventas_por_producto = (
        datos_combinados.groupby(['Tienda', 'Producto'])
        .agg(Ventas=('Cantidad de cuotas', 'sum'))
        .reset_index()
    )

    # Obtener los productos más y menos vendidos
    mas_vendidos = (
        ventas_por_producto.sort_values(by=['Tienda', 'Ventas'], ascending=[True, False])
        .groupby('Tienda')
        .head(top_n)
    )
    menos_vendidos = (
        ventas_por_producto.sort_values(by=['Tienda', 'Ventas'], ascending=[True, True])
        .groupby('Tienda')
        .head(top_n)
    )

    return {'Mas Vendidos': mas_vendidos, 'Menos Vendidos': menos_vendidos}

# Función para identificar los productos con mayores y menores ingresos
def identificar_productos_mayores_y_menores_ingresos(datos_combinados, top_n=3):
    """
    Identifica los productos que generaron mayores y menores ingresos en cada tienda.

    Parámetros:
        datos_combinados (pd.DataFrame): El DataFrame combinado de todas las tiendas.
        top_n (int): Número de productos a incluir en los resultados.

    Retorna:
        dict: Un diccionario con los productos de mayores y menores ingresos por tienda.
    """
    # Calcular los ingresos totales por producto (Precio * Ventas)
    ingresos_por_producto = (
        datos_combinados.assign(Ingresos=lambda x: x['Precio'] * x['Cantidad de cuotas'])
        .groupby(['Tienda', 'Producto'])['Ingresos']
        .sum()
        .reset_index()
    )

    # Obtener los productos con mayores y menores ingresos
    mayores_ingresos = (
        ingresos_por_producto.sort_values(by=['Tienda', 'Ingresos'], ascending=[True, False])
        .groupby('Tienda')
        .head(top_n)
    )
    menores_ingresos = (
        ingresos_por_producto.sort_values(by=['Tienda', 'Ingresos'], ascending=[True, True])
        .groupby('Tienda')
        .head(top_n)
    )

    return {'Mayores Ingresos': mayores_ingresos, 'Menores Ingresos': menores_ingresos}

# Función principal
def main():
    # URLs de los archivos CSV
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv    "
    ]

    # Cargar los datos de cada tienda y agregar la columna 'Tienda'
    dataframes = []
    for i, url in enumerate(urls, start=1):
        df = leer_csv_pandas(url)
        if df is not None:
            df = agregar_columna_tienda(df, f"Tienda {i}")
            dataframes.append(df)

    # Combinar todos los DataFrames en uno solo
    datos_combinados = pd.concat(dataframes, ignore_index=True)

    # Paso 1: Calcular las ventas por categoría
    ventas_por_categoria = calcular_ventas_por_categoria(datos_combinados)
    print("\nVentas por categoría en cada tienda:")
    for tienda, grupo in ventas_por_categoria.groupby('Tienda'):
        print(f"\n{tienda}:")
        for _, row in grupo.iterrows():
            print(f"  - {row['Categoría del Producto']}: {int(row['Ventas'])} unidades")

    # Paso 2: Identificar las categorías más populares
    categorias_mas_populares = identificar_categorias_mas_populares(ventas_por_categoria)
    print("\nCategorías más populares de cada tienda:")
    for _, row in categorias_mas_populares.iterrows():
        print(f"{row['Tienda']}: {row['Categoría del Producto']} ({int(row['Ventas'])} unidades)")

    # Paso 3: Identificar los productos más y menos vendidos
    resultados_top_productos = identificar_top_productos(datos_combinados)
    print("\nProductos más vendidos y menos vendidos por tienda:")
    for tienda, grupo in resultados_top_productos['Mas Vendidos'].groupby('Tienda'):
        print(f"\n{tienda}:")
        print("Productos más vendidos:")
        for _, row in grupo.iterrows():
            print(f"  - {row['Producto']}: {int(row['Ventas'])} unidades")
    for tienda, grupo in resultados_top_productos['Menos Vendidos'].groupby('Tienda'):
        print("Productos menos vendidos:")
        for _, row in grupo.iterrows():
            print(f"  - {row['Producto']}: {int(row['Ventas'])} unidades")

    # Paso 4: Identificar los productos con mayores y menores ingresos
    productos_ingresos = identificar_productos_mayores_y_menores_ingresos(datos_combinados)
    print("\nProductos que generaron mayores ingresos por tienda:")
    for tienda, grupo in productos_ingresos['Mayores Ingresos'].groupby('Tienda'):
        print(f"\n{tienda}:")
        print("Productos con mayores ingresos:")
        for _, row in grupo.iterrows():
            print(f"  - {row['Producto']}: ${int(row['Ingresos'])}")
    for tienda, grupo in productos_ingresos['Menores Ingresos'].groupby('Tienda'):
        print("Productos con menores ingresos:")
        for _, row in grupo.iterrows():
            print(f"  - {row['Producto']}: ${int(row['Ingresos'])}")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import csv
from urllib.request import urlopen

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url.strip()) as respuesta:  # Eliminamos espacios adicionales en la URL
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para agregar una columna 'Tienda' a cada lista de datos
def agregar_columna_tienda(datos, nombre_tienda):
    """
    Agrega una clave 'Tienda' a cada diccionario en la lista de datos.

    Parámetros:
        datos (list): Lista de diccionarios que representan los datos.
        nombre_tienda (str): El nombre de la tienda.

    Retorna:
        list: La lista actualizada con la clave 'Tienda'.
    """
    for fila in datos:
        fila['Tienda'] = nombre_tienda
    return datos

# Función para combinar todos los datos en una sola lista
def combinar_datos(*listas_datos):
    """
    Combina múltiples listas de datos en una sola lista.

    Parámetros:
        *listas_datos: Varias listas de datos.

    Retorna:
        list: Una lista combinada de todas las listas de entrada.
    """
    return [fila for lista in listas_datos for fila in lista]

# Función para calcular las ventas por categoría
def calcular_ventas_por_categoria(datos_combinados):
    """
    Calcula la cantidad de productos vendidos por categoría en cada tienda.

    Parámetros:
        datos_combinados (list): Lista combinada de datos de todas las tiendas.

    Retorna:
        dict: Un diccionario con las ventas por categoría.
    """
    ventas_por_categoria = {}
    for fila in datos_combinados:
        tienda = fila['Tienda']
        categoria = fila['Categoría del Producto']
        if tienda not in ventas_por_categoria:
            ventas_por_categoria[tienda] = {}
        if categoria not in ventas_por_categoria[tienda]:
            ventas_por_categoria[tienda][categoria] = 0
        ventas_por_categoria[tienda][categoria] += int(fila['Cantidad de cuotas'])
    return ventas_por_categoria

# Función para identificar las categorías más populares
def identificar_categorias_mas_populares(ventas_por_categoria):
    """
    Identifica las categorías más populares de cada tienda.

    Parámetros:
        ventas_por_categoria (dict): Diccionario con las ventas por categoría.

    Retorna:
        dict: Un diccionario con las categorías más populares.
    """
    categorias_mas_populares = {}
    for tienda, categorias in ventas_por_categoria.items():
        categoria_mas_popular = max(categorias, key=categorias.get)
        categorias_mas_populares[tienda] = {
            'Categoría': categoria_mas_popular,
            'Ventas': categorias[categoria_mas_popular]
        }
    return categorias_mas_populares

# Función para identificar los productos más y menos vendidos
def identificar_top_productos(datos_combinados, top_n=3):
    """
    Identifica los productos más vendidos y menos vendidos de cada tienda.

    Parámetros:
        datos_combinados (list): Lista combinada de datos de todas las tiendas.
        top_n (int): Número de productos a incluir en los resultados.

    Retorna:
        dict: Un diccionario con los productos más y menos vendidos por tienda.
    """
    ventas_por_producto = {}
    for fila in datos_combinados:
        tienda = fila['Tienda']
        producto = fila['Producto']
        if tienda not in ventas_por_producto:
            ventas_por_producto[tienda] = {}
        if producto not in ventas_por_producto[tienda]:
            ventas_por_producto[tienda][producto] = 0
        ventas_por_producto[tienda][producto] += int(fila['Cantidad de cuotas'])

    resultados = {}
    for tienda, productos in ventas_por_producto.items():
        mas_vendidos = sorted(productos.items(), key=lambda x: x[1], reverse=True)[:top_n]
        menos_vendidos = sorted(productos.items(), key=lambda x: x[1])[:top_n]
        resultados[tienda] = {
            'Mas Vendidos': mas_vendidos,
            'Menos Vendidos': menos_vendidos
        }
    return resultados

# Función para identificar los productos con mayores y menores ingresos
def identificar_productos_mayores_y_menores_ingresos(datos_combinados, top_n=3):
    """
    Identifica los productos que generaron mayores y menores ingresos en cada tienda.

    Parámetros:
        datos_combinados (list): Lista combinada de datos de todas las tiendas.
        top_n (int): Número de productos a incluir en los resultados.

    Retorna:
        dict: Un diccionario con los productos de mayores y menores ingresos por tienda.
    """
    ingresos_por_producto = {}
    for fila in datos_combinados:
        tienda = fila['Tienda']
        producto = fila['Producto']
        precio = float(fila['Precio'])
        cantidad = int(fila['Cantidad de cuotas'])
        ingreso = precio * cantidad
        if tienda not in ingresos_por_producto:
            ingresos_por_producto[tienda] = {}
        if producto not in ingresos_por_producto[tienda]:
            ingresos_por_producto[tienda][producto] = 0
        ingresos_por_producto[tienda][producto] += ingreso

    resultados = {}
    for tienda, productos in ingresos_por_producto.items():
        mayores_ingresos = sorted(productos.items(), key=lambda x: x[1], reverse=True)[:top_n]
        menores_ingresos = sorted(productos.items(), key=lambda x: x[1])[:top_n]
        resultados[tienda] = {
            'Mayores Ingresos': mayores_ingresos,
            'Menores Ingresos': menores_ingresos
        }
    return resultados

# Función principal
def main():
    # URLs de los archivos CSV
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv    "
    ]

    # Cargar los datos de cada tienda y agregar la columna 'Tienda'
    dataframes = []
    for i, url in enumerate(urls, start=1):
        datos = leer_csv(url)
        if datos:
            datos = agregar_columna_tienda(datos, f"Tienda {i}")
            dataframes.append(datos)

    # Combinar todos los datos en una sola lista
    datos_combinados = combinar_datos(*dataframes)

    # Paso 1: Calcular las ventas por categoría
    ventas_por_categoria = calcular_ventas_por_categoria(datos_combinados)
    print("\nVentas por categoría en cada tienda:")
    for tienda, categorias in ventas_por_categoria.items():
        print(f"\n{tienda}:")
        for categoria, ventas in categorias.items():
            print(f"  - {categoria}: {ventas} unidades")

    # Paso 2: Identificar las categorías más populares
    categorias_mas_populares = identificar_categorias_mas_populares(ventas_por_categoria)
    print("\nCategorías más populares de cada tienda:")
    for tienda, datos in categorias_mas_populares.items():
        print(f"{tienda}: {datos['Categoría']} ({datos['Ventas']} unidades)")

    # Paso 3: Identificar los productos más y menos vendidos
    resultados_top_productos = identificar_top_productos(datos_combinados)
    print("\nProductos más vendidos y menos vendidos por tienda:")
    for tienda, datos in resultados_top_productos.items():
        print(f"\n{tienda}:")
        print("Productos más vendidos:")
        for producto, ventas in datos['Mas Vendidos']:
            print(f"  - {producto}: {ventas} unidades")
        print("Productos menos vendidos:")
        for producto, ventas in datos['Menos Vendidos']:
            print(f"  - {producto}: {ventas} unidades")

    # Paso 4: Identificar los productos con mayores y menores ingresos
    productos_ingresos = identificar_productos_mayores_y_menores_ingresos(datos_combinados)
    print("\nProductos que generaron mayores ingresos por tienda:")
    for tienda, datos in productos_ingresos.items():
        print(f"\n{tienda}:")
        print("Productos con mayores ingresos:")
        for producto, ingresos in datos['Mayores Ingresos']:
            print(f"  - {producto}: ${ingresos:.2f}")
        print("Productos con menores ingresos:")
        for producto, ingresos in datos['Menores Ingresos']:
            print(f"  - {producto}: ${ingresos:.2f}")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import csv
from urllib.request import urlopen

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url.strip()) as respuesta:  # Eliminamos espacios adicionales en la URL
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para agregar una columna 'Tienda' a cada lista de datos
def agregar_columna_tienda(datos, nombre_tienda):
    """
    Agrega una clave 'Tienda' a cada diccionario en la lista de datos.

    Parámetros:
        datos (list): Lista de diccionarios que representan los datos.
        nombre_tienda (str): El nombre de la tienda.

    Retorna:
        list: La lista actualizada con la clave 'Tienda'.
    """
    for fila in datos:
        fila['Tienda'] = nombre_tienda
    return datos

# Función para combinar todos los datos en una sola lista
def combinar_datos(*listas_datos):
    """
    Combina múltiples listas de datos en una sola lista.

    Parámetros:
        *listas_datos: Varias listas de datos.

    Retorna:
        list: Una lista combinada de todas las listas de entrada.
    """
    return [fila for lista in listas_datos for fila in lista]

# Función para calcular el costo de envío promedio por tienda
def calcular_costo_envio_promedio(datos_combinados):
    """
    Calcula el costo de envío promedio para cada tienda.

    Parámetros:
        datos_combinados (list): Lista combinada de datos de todas las tiendas.

    Retorna:
        dict: Un diccionario con el costo de envío promedio por tienda.
    """
    costos_por_tienda = {}
    for fila in datos_combinados:
        tienda = fila['Tienda']
        costo_envio = float(fila.get('Costo de envío', 0))  # Convertir a float, 0 si falta
        if tienda not in costos_por_tienda:
            costos_por_tienda[tienda] = {'total': 0, 'cantidad': 0}
        costos_por_tienda[tienda]['total'] += costo_envio
        costos_por_tienda[tienda]['cantidad'] += 1

    # Calcular el promedio
    costo_envio_promedio = {}
    for tienda, datos in costos_por_tienda.items():
        promedio = datos['total'] / datos['cantidad'] if datos['cantidad'] > 0 else 0
        costo_envio_promedio[tienda] = promedio

    return costo_envio_promedio

# Función principal
def main():
    # URLs de los archivos CSV
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv    "
    ]

    # Cargar los datos de cada tienda y agregar la columna 'Tienda'
    dataframes = []
    for i, url in enumerate(urls, start=1):
        datos = leer_csv(url)
        if datos:
            datos = agregar_columna_tienda(datos, f"Tienda {i}")
            dataframes.append(datos)

    # Combinar todos los datos en una sola lista
    datos_combinados = combinar_datos(*dataframes)

    # Paso: Calcular el costo de envío promedio por tienda
    costo_envio_promedio = calcular_costo_envio_promedio(datos_combinados)

    # Mostrar los resultados en el formato solicitado
    print("\nCosto de Envío Promedio por Tienda:")
    print(f"{'Tienda':<10} {'Costo de Envío Promedio':>25}")
    for tienda, promedio in costo_envio_promedio.items():
        print(f"{tienda:<10} {promedio:>25.6f}")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Función para leer un archivo CSV desde una URL usando Pandas
def leer_csv_pandas(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en un DataFrame de Pandas.
    """
    try:
        df = pd.read_csv(url.strip())  # Eliminamos espacios adicionales en la URL
        print(f"Datos cargados exitosamente desde {url}")
        return df
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return None

# Función para agregar una columna 'Tienda' a cada DataFrame
def agregar_columna_tienda(df, nombre_tienda):
    """
    Agrega una columna 'Tienda' al DataFrame indicando el nombre de la tienda.

    Parámetros:
        df (pd.DataFrame): El DataFrame original.
        nombre_tienda (str): El nombre de la tienda.

    Retorna:
        pd.DataFrame: El DataFrame con la columna 'Tienda' agregada.
    """
    df['Tienda'] = nombre_tienda
    return df

# Función para combinar todos los DataFrames en uno solo
def combinar_datos(*dataframes):
    """
    Combina múltiples DataFrames en uno solo.

    Parámetros:
        *dataframes: Varias listas de datos.

    Retorna:
        pd.DataFrame: Un DataFrame combinado de todas las listas de entrada.
    """
    return pd.concat(dataframes, ignore_index=True)

# Función para calcular los ingresos totales por tienda
def calcular_ingresos_totales(datos_combinados):
    """
    Calcula los ingresos totales por tienda.

    Parámetros:
        datos_combinados (pd.DataFrame): El DataFrame combinado de todas las tiendas.

    Retorna:
        pd.DataFrame: Un DataFrame con los ingresos totales por tienda.
    """
    ingresos_por_tienda = (
        datos_combinados.assign(Ingresos=lambda x: x['Precio'] * x['Cantidad de cuotas'])
        .groupby('Tienda')['Ingresos']
        .sum()
        .reset_index()
    )
    return ingresos_por_tienda

# Función para calcular la participación de cada tienda en el total de ingresos
def calcular_participacion_ingresos(ingresos_totales):
    """
    Calcula la participación de cada tienda en el total de ingresos.

    Parámetros:
        ingresos_totales (pd.DataFrame): El DataFrame con los ingresos totales por tienda.

    Retorna:
        pd.DataFrame: Un DataFrame con la participación de cada tienda.
    """
    total_ingresos = ingresos_totales['Ingresos'].sum()
    participacion = ingresos_totales.copy()
    participacion['Participación (%)'] = (participacion['Ingresos'] / total_ingresos) * 100
    return participacion

# Función para calcular el ingreso promedio por venta por tienda
def calcular_ingreso_promedio_por_venta(datos_combinados):
    """
    Calcula el ingreso promedio por venta para cada tienda.

    Parámetros:
        datos_combinados (pd.DataFrame): El DataFrame combinado de todas las tiendas.

    Retorna:
        pd.DataFrame: Un DataFrame con el ingreso promedio por venta por tienda.
    """
    ingreso_promedio_por_tienda = (
        datos_combinados.assign(Ingresos=lambda x: x['Precio'] * x['Cantidad de cuotas'])
        .groupby('Tienda')['Ingresos']
        .mean()
        .reset_index(name='Ingreso Promedio por Venta')
    )
    return ingreso_promedio_por_tienda

# Función principal
def main():
    # URLs de los archivos CSV
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv    "
    ]

    # Cargar los datos de cada tienda y agregar la columna 'Tienda'
    dataframes = []
    for i, url in enumerate(urls, start=1):
        df = leer_csv_pandas(url)
        if df is not None:
            df = agregar_columna_tienda(df, f"Tienda {i}")
            dataframes.append(df)

    # Combinar todos los DataFrames en uno solo
    datos_combinados = combinar_datos(*dataframes)

    # Paso 1: Calcular los ingresos totales por tienda
    ingresos_totales = calcular_ingresos_totales(datos_combinados)

    # Paso 2: Calcular la participación de cada tienda en el total de ingresos
    participacion_ingresos = calcular_participacion_ingresos(ingresos_totales)

    # Paso 3: Calcular el ingreso promedio por venta por tienda
    ingreso_promedio_por_venta = calcular_ingreso_promedio_por_venta(datos_combinados)

    # Gráfico 1: Ingresos Totales por Tienda (Barras)
    plt.figure(figsize=(10, 6))
    plt.bar(ingresos_totales['Tienda'], ingresos_totales['Ingresos'], color='skyblue', edgecolor='black')
    plt.title("Ingresos Totales por Tienda", fontsize=16)
    plt.xlabel("Tienda", fontsize=12)
    plt.ylabel("Ingresos Totales ($)", fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Gráfico 2: Participación de Cada Tienda en el Total de Ingresos (Pastel)
    plt.figure(figsize=(8, 8))
    plt.pie(
        participacion_ingresos['Participación (%)'],
        labels=participacion_ingresos['Tienda'],
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors
    )
    plt.title("Participación de Cada Tienda en el Total de Ingresos", fontsize=16)
    plt.tight_layout()
    plt.show()

    # Gráfico 3: Ingreso Promedio por Venta por Tienda (Líneas)
    plt.figure(figsize=(10, 6))
    plt.plot(
        ingreso_promedio_por_venta['Tienda'],
        ingreso_promedio_por_venta['Ingreso Promedio por Venta'],
        marker='o',
        color='green',
        linestyle='-',
        linewidth=2,
        markersize=8
    )
    plt.title("Ingreso Promedio por Venta por Tienda", fontsize=16)
    plt.xlabel("Tienda", fontsize=12)
    plt.ylabel("Ingreso Promedio por Venta ($)", fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
