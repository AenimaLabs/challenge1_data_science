```python
import pandas as pd

# URLs de los archivos CSV
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv "
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv "
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv "
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv "

# Cargar los datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Función para calcular el ingreso total
def calcular_ingreso_total(dataframe, columna_precio="Precio"):
    """
    Calcula el ingreso total sumando los valores de la columna especificada.

    Parámetros:
    - dataframe: DataFrame de pandas que contiene los datos.
    - columna_precio: Nombre de la columna que contiene los precios (por defecto es 'Precio').

    Retorna:
    - Ingreso total como un número flotante.
    """
    if columna_precio in dataframe.columns:
        return dataframe[columna_precio].sum()
    else:
        raise ValueError(f"La columna '{columna_precio}' no existe en el DataFrame.")

# Calcular el ingreso total para cada tienda usando la función
ingreso_tienda1 = calcular_ingreso_total(tienda)
ingreso_tienda2 = calcular_ingreso_total(tienda2)
ingreso_tienda3 = calcular_ingreso_total(tienda3)
ingreso_tienda4 = calcular_ingreso_total(tienda4)

# Mostrar los resultados
print(f"Ingreso total Tienda 1: ${ingreso_tienda1:.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda2:.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda3:.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda4:.2f}")
```

```python
import pandas as pd


# Función para leer un archivo CSV desde una URL usando Pandas
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en un DataFrame de Pandas.
    """
    try:
        df = pd.read_csv(url)
        print(f"Datos cargados exitosamente desde {url}")
        return df
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return pd.DataFrame()  # Retorna un DataFrame vacío en caso de error

# Función para calcular el ingreso total de una tienda
def calcular_ingreso_total(df, columna_precio="Precio"):
    """
    Calcula el ingreso total sumando los valores de la columna de precios.

    Parámetros:
        df (pd.DataFrame): El DataFrame que contiene los datos de la tienda.
        columna_precio (str): El nombre de la columna que contiene los precios.

    Retorna:
        float: El ingreso total calculado.
    """
    if columna_precio not in df.columns:
        print(f"La columna '{columna_precio}' no existe en el DataFrame.")
        return None

    ingreso_total = df[columna_precio].sum()
    return ingreso_total

#Función para calcular la sumatoria de la facturación de todas las tiendas
def calcular_sumatoria_facturacion(ingresos_totales):
    """
    Calcula la sumatoria de los ingresos totales de todas las tiendas.

    Parámetros:
        ingresos_totales (dict): Un diccionario donde las claves son los nombres de las tiendas
                                 y los valores son los ingresos totales de cada tienda.

    Retorna:
        float: La sumatoria de los ingresos totales de todas las tiendas.
    """
    if not ingresos_totales:
        print("No hay datos de ingresos disponibles.")
        return None

    sumatoria = sum(ingresos_totales.values())
    return sumatoria

# URLs de los archivos CSV
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv "
]

# Calcular el ingreso total para cada tienda
ingresos_totales = {}
for i, url in enumerate(urls, start=1):
    # Leer el CSV
    tienda_df = leer_csv(url)
    if tienda_df is not None:
        # Calcular el ingreso total
        ingreso_total = calcular_ingreso_total(tienda_df, columna_precio="Precio")
        if ingreso_total is not None:
            ingresos_totales[f"Tienda {i}"] = ingreso_total

# Mostrar los ingresos totales de cada tienda
print("Ingresos Totales por Tienda:")
for tienda, ingreso in ingresos_totales.items():
    print(f"{tienda}: ${ingreso:.2f}")

# Calcular la sumatoria de la facturación de todas las tiendas
sumatoria_facturacion = calcular_sumatoria_facturacion(ingresos_totales)
if sumatoria_facturacion is not None:
    print(f"\nFacturación Total de Todas las Tiendas: ${sumatoria_facturacion:.2f}")
```

```python
#utilizando Python estándar
import csv
import urllib.request

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url) as respuesta:
            # Decodificar los datos
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para calcular el ingreso total de una tienda
def calcular_ingreso_total(url):
    ingreso_total = 0
    with urllib.request.urlopen(url) as response:
        lines = [line.decode('utf-8') for line in response]
        reader = csv.DictReader(lines)
        for row in reader:
            # Convertir el precio a float y sumarlo al ingreso total
            precio = float(row['Precio'])
            ingreso_total += precio
    return ingreso_total

# URLs de las tiendas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv "
]

# Calcular el ingreso total para cada tienda
ingresos_totales = []
for i, url in enumerate(urls, start=1):
    ingreso_total = calcular_ingreso_total(url)
    ingresos_totales.append(ingreso_total)
    print(f"Ingreso total de la Tienda {i}: ${ingreso_total:.2f}")

# Resultado final
print("\nResumen de ingresos totales:")
for i, ingreso in enumerate(ingresos_totales, start=1):
    print(f"Tienda {i}: ${ingreso:.2f}")
```

```python
import csv
from urllib.request import urlopen

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url) as respuesta:
            # Decodificar los datos
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para calcular el ingreso total de una tienda
def calcular_ingreso_total(data, columna_precio='Precio'):
    """
    Calcula el ingreso total sumando los valores de la columna 'Precio' de los datos.

    Args:
    - data: Lista de diccionarios que contiene los datos de la tienda.
    - columna_precio: Nombre de la columna que contiene los precios (por defecto es 'Precio').

    Returns:
    - Ingreso total como un número flotante.
    """
    ingreso_total = 0
    for row in data:
        try:
            # Convertir el precio a float y sumarlo al ingreso total
            precio = float(row[columna_precio])
            ingreso_total += precio
        except KeyError:
            print(f"Error: La columna '{columna_precio}' no existe en los datos.")
            return None
        except ValueError:
            print(f"Error: El valor '{row[columna_precio]}' no es un número válido.")
            return None
    return ingreso_total

# URLs de las tiendas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv ",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv "
]

# Calcular el ingreso total para cada tienda
ingresos_totales = []
for i, url in enumerate(urls, start=1):
    # Leer los datos desde la URL
    data = leer_csv(url)
    if data:
        # Calcular el ingreso total
        ingreso_total = calcular_ingreso_total(data)
        if ingreso_total is not None:
            ingresos_totales.append(ingreso_total)
            print(f"Ingreso total de la Tienda {i}: ${ingreso_total:.2f}")
        else:
            print(f"No se pudo calcular el ingreso total para la Tienda {i}.")
    else:
        print(f"No se pudieron leer los datos de la Tienda {i}.")

# Resultado final
print("\nResumen de ingresos totales:")
for i, ingreso in enumerate(ingresos_totales, start=1):
    print(f"Tienda {i}: ${ingreso:.2f}")
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Agregar columna Tienda
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar datos
datos_combinados = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Contar productos vendidos por categoría en cada tienda
productos_por_categoria = datos_combinados.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad Vendida')

# Encontrar las categorías más populares por tienda
categorias_populares = productos_por_categoria.loc[productos_por_categoria.groupby('Tienda')['Cantidad Vendida'].idxmax()]

# Mostrar resultados
print("Categorías más populares por tienda:")
print(categorias_populares)

print("\nResumen de productos vendidos por categoría en cada tienda:")
print(productos_por_categoria)
```

```python
import pandas as pd

# URLs de los archivos CSV
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv "
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv "
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv "
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv "

# Cargar los datos de cada tienda
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Agregar una columna 'Tienda' a cada DataFrame
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar todos los DataFrames en uno solo
datos_combinados = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Paso 1: Calcular la cantidad de productos vendidos por categoría en cada tienda
ventas_por_categoria = (
    datos_combinados.groupby(['Tienda', 'Categoría del Producto'])
    .size()
    .reset_index(name='Ventas')
)

print("Ventas por categoría en cada tienda:")
print(ventas_por_categoria)

# Paso 2: Identificar las categorías más populares de cada tienda
categorias_mas_populares = (
    ventas_por_categoria.loc[ventas_por_categoria.groupby('Tienda')['Ventas'].idxmax()]
)

print("\nCategorías más populares de cada tienda:")
print(categorias_mas_populares)

# Paso 3: Identificar los 3 artículos más vendidos y los 3 menos vendidos de cada tienda
# Agrupar por Tienda y Producto, y contar el número de ventas
ventas_por_producto = (
    datos_combinados.groupby(['Tienda', 'Producto'])
    .size()
    .reset_index(name='Ventas')
)

# Función para obtener los 3 productos más vendidos y menos vendidos de cada tienda
def top_productos(df, top_n=3):
    mas_vendidos = df.nlargest(top_n, 'Ventas')
    menos_vendidos = df.nsmallest(top_n, 'Ventas')
    return mas_vendidos, menos_vendidos

# Aplicar la función a cada tienda
resultados = {}
for tienda in datos_combinados['Tienda'].unique():
    df_tienda = ventas_por_producto[ventas_por_producto['Tienda'] == tienda]
    mas_vendidos, menos_vendidos = top_productos(df_tienda)
    resultados[tienda] = {'Mas Vendidos': mas_vendidos, 'Menos Vendidos': menos_vendidos}

print("\nProductos más vendidos y menos vendidos por tienda:")
for tienda, datos in resultados.items():
    print(f"\n{tienda}:")
    print("Productos más vendidos:")
    print(datos['Mas Vendidos'])
    print("Productos menos vendidos:")
    print(datos['Menos Vendidos'])

# Paso 4: Identificar los productos que generaron mayores ingresos
# Calcular los ingresos totales por producto (Precio * Ventas)
ingresos_por_producto = (
    datos_combinados.assign(Ingresos=lambda x: x['Precio'] * x['Cantidad de cuotas'])
    .groupby(['Tienda', 'Producto'])['Ingresos']
    .sum()
    .reset_index()
)

# Obtener los productos con mayores ingresos por tienda
productos_mayores_ingresos = {}
for tienda in datos_combinados['Tienda'].unique():
    df_tienda = ingresos_por_producto[ingresos_por_producto['Tienda'] == tienda]
    productos_mayores_ingresos[tienda] = df_tienda.nlargest(3, 'Ingresos')

print("\nProductos que generaron mayores ingresos por tienda:")
for tienda, datos in productos_mayores_ingresos.items():
    print(f"\n{tienda}:")
    print(datos)
```

```python
import pandas as pd

# Función para leer un archivo CSV desde una URL usando Pandas
def leer_csv_pandas(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en un DataFrame de Pandas.
    """
    try:
        df = pd.read_csv(url)
        print(f"Datos cargados exitosamente desde {url}")
        return df
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return pd.DataFrame()  # Retorna un DataFrame vacío en caso de error

# Función para contar productos vendidos por categoría
def contar_productos_por_categoria(df):
    """
    Recibe un DataFrame y cuenta los productos vendidos por categoría.
    Retorna un diccionario con las categorías como claves y los conteos como valores.
    """
    if "Categoría del Producto" not in df.columns:
        print("La columna 'Categoría del Producto' no está presente en los datos.")
        return {}
    conteo_categorias = df["Categoría del Producto"].value_counts().to_dict()
    return conteo_categorias

# Función para calcular el total de ventas en dinero por categorías
def calcular_ventas_por_categorias(df):
    """
    Calcula el total de ventas en dinero por categorías.
    Retorna un diccionario con las categorías como claves y los totales de ventas como valores.
    """
    try:
        # Asegurarse de que las columnas necesarias existan
        if "Categoría del Producto" not in df.columns or "Precio" not in df.columns or "Cantidad de cuotas" not in df.columns:
            print("Faltan columnas necesarias para calcular las ventas por categorías.")
            return {}

        # Calcular el monto de venta por fila
        df["Monto Venta"] = df["Precio"] * df["Cantidad de cuotas"]

        # Agrupar por categoría y sumar los montos de venta
        ventas_por_categorias = df.groupby("Categoría del Producto")["Monto Venta"].sum().to_dict()
        return ventas_por_categorias
    except Exception as e:
        print(f"Error al calcular las ventas por categorías: {e}")
        return {}

# Función para calcular el total de ventas en dinero
def calcular_total_ventas(df):
    """
    Calcula el total de ventas en dinero sumando el precio de cada producto multiplicado por la cantidad de cuotas.
    Retorna el total de ventas.
    """
    try:
        if "Precio" not in df.columns or "Cantidad de cuotas" not in df.columns:
            print("Faltan columnas necesarias para calcular el total de ventas.")
            return 0
        total_ventas = (df["Precio"] * df["Cantidad de cuotas"]).sum()
        return total_ventas
    except Exception as e:
        print(f"Error al calcular el total de ventas: {e}")
        return 0

# Función para mostrar los resultados
def mostrar_resultados(resultados, totales_ventas, ventas_por_categorias):
    """
    Muestra los conteos de productos por categoría, el total de ventas en dinero,
    y el total de ventas por categorías.
    """
    for tienda, conteo in resultados.items():
        print(f"\n{tienda}:")
        if not conteo:  # Si el conteo está vacío
            print("  - No se encontraron datos de ventas.")
        else:
            for categoria, cantidad in conteo.items():
                print(f"  - {categoria}: {cantidad} productos vendidos")

        # Mostrar el total de ventas por categorías
        print("  - Ventas por categorías:")
        for categoria, total_venta in ventas_por_categorias[tienda].items():
            print(f"    - {categoria}: ${total_venta:,.2f}")

        # Mostrar el total de ventas general
        print(f"  - Total de ventas: ${totales_ventas[tienda]:,.2f}")

# Función principal
def main():
    # Opción 1: Usar URLs (comentar si se usan archivos locales)
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv    ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv    "
    ]

    # Procesar los datos de todas las tiendas
    resultados = {}
    totales_ventas = {}
    ventas_por_categorias = {}
    for i, url in enumerate(urls, start=1):  # Cambiar `urls` por `rutas_locales` si se usan archivos locales
        nombre_tienda = f"Tienda {i}"

        # Leer los datos de la tienda
        df_tienda = leer_csv_pandas(url.strip())  # Eliminar espacios adicionales en la URL

        if df_tienda.empty:  # Si el DataFrame está vacío, saltar esta tienda
            print(f"No se pudo cargar datos para {nombre_tienda}.")
            continue

        # Contar productos por categoría
        conteo_categorias = contar_productos_por_categoria(df_tienda)
        resultados[nombre_tienda] = conteo_categorias

        # Calcular el total de ventas en dinero
        total_venta = calcular_total_ventas(df_tienda)
        totales_ventas[nombre_tienda] = total_venta

        # Calcular el total de ventas por categorías
        ventas_por_categoria = calcular_ventas_por_categorias(df_tienda)
        ventas_por_categorias[nombre_tienda] = ventas_por_categoria

    # Mostrar los resultados
    mostrar_resultados(resultados, totales_ventas, ventas_por_categorias)

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
#utilizando Python estándar
import csv
from urllib.request import urlopen

# Función para leer un archivo CSV desde una URL
def leer_csv(url):
    """
    Lee un archivo CSV desde una URL y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with urlopen(url) as respuesta:
            # Decodificar los datos
            lineas = respuesta.read().decode('utf-8').splitlines()
            reader = csv.DictReader(lineas)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {url}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {url}: {e}")
        return []

# Función para leer un archivo CSV local (en este caso no la usamos)
def leer_csv_local(ruta_archivo):
    """
    Lee un archivo CSV local y lo convierte en una lista de diccionarios.
    """
    try:
        datos = []
        with open(ruta_archivo, mode='r', encoding='utf-8') as archivo:
            reader = csv.DictReader(archivo)
            for fila in reader:
                datos.append(fila)
        print(f"Datos cargados exitosamente desde {ruta_archivo}")
        return datos
    except Exception as e:
        print(f"Error al cargar datos desde {ruta_archivo}: {e}")
        return []

# Función para contar productos vendidos por categoría
def contar_productos_por_categoria(datos):
    """
    Recibe una lista de datos (diccionarios) y cuenta los productos vendidos por categoría.
    Retorna un diccionario con las categorías como claves y los conteos como valores.
    """
    conteo_categorias = {}
    for fila in datos:
        if "Categoría del Producto" not in fila:
            print("La columna 'Categoría del Producto' no está presente en los datos.")
            return {}
        categoria = fila["Categoría del Producto"]
        if categoria not in conteo_categorias:
            conteo_categorias[categoria] = 0
        conteo_categorias[categoria] += 1
    return conteo_categorias

# Función para calcular el total de ventas en dinero por categorías
def calcular_ventas_por_categorias(datos):
    """
    Calcula el total de ventas en dinero por categorías.
    Retorna un diccionario con las categorías como claves y los totales de ventas como valores.
    """
    ventas_por_categorias = {}
    errores = 0
    for fila in datos:
        try:
            categoria = fila["Categoría del Producto"]
            precio = float(fila["Precio"])
            cantidad = int(fila["Cantidad de cuotas"])
            monto_venta = precio * cantidad
            if categoria not in ventas_por_categorias:
                ventas_por_categorias[categoria] = 0
            ventas_por_categorias[categoria] += monto_venta
        except (ValueError, KeyError):
            errores += 1
            print("Error al procesar el precio o la categoría de un producto. Se omitirá.")
    if errores > 0:
        print(f"Se omitieron {errores} filas debido a errores.")
    return ventas_por_categorias

# Función para calcular el total de ventas en dinero
def calcular_total_ventas(datos):
    """
    Calcula el total de ventas en dinero sumando el precio de cada producto.
    Retorna el total de ventas.
    """
    total_ventas = 0
    for fila in datos:
        try:
            precio = float(fila["Precio"])
            total_ventas += precio
        except (ValueError, KeyError):
            print("Error al procesar el precio de un producto. Se omitirá.")
    return total_ventas

# Función para mostrar los resultados
def mostrar_resultados(resultados, totales_ventas, ventas_por_categorias):
    """
    Muestra los conteos de productos por categoría, el total de ventas en dinero,
    y el total de ventas por categorías.
    """
    for tienda, conteo in resultados.items():
        print(f"\n{tienda}:")
        if not conteo:  # Si el conteo está vacío
            print("  - No se encontraron datos de ventas.")
        else:
            for categoria, cantidad in conteo.items():
                print(f"  - {categoria}: {cantidad} productos vendidos")

        # Mostrar el total de ventas por categorías
        print("  - Ventas por categorías:")
        for categoria, total_venta in ventas_por_categorias[tienda].items():
            print(f"    - {categoria}: ${total_venta:,.2f}")

        # Mostrar el total de ventas general
        print(f"  - Total de ventas: ${totales_ventas[tienda]:,.2f}")

# Función principal
def main():
    # Opción 1: Usar URLs (comentada si prefieres usar archivos locales)
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv ",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv "
    ]

    # Opción 2: Usar archivos locales (descomenta si prefieres usar archivos locales)
    # rutas_locales = [
    #     "tienda_1.csv",
    #     "tienda_2.csv",
    #     "tienda_3.csv",
    #     "tienda_4.csv"
    # ]

    # Procesar los datos de todas las tiendas
    resultados = {}
    totales_ventas = {}
    ventas_por_categorias = {}
    for i, url in enumerate(urls, start=1):  # Cambia `urls` por `rutas_locales` si usas archivos locales
        nombre_tienda = f"Tienda {i}"

        # Descomenta la línea correspondiente según la opción que uses
        datos_tienda = leer_csv(url)  # Para URLs
        # datos_tienda = leer_csv_local(rutas_locales[i-1])  # Para archivos locales

        if not datos_tienda:  # Si los datos están vacíos, saltar esta tienda
            print(f"No se pudo cargar datos para {nombre_tienda}.")
            continue
         # Contar productos por categoría
        conteo_categorias = contar_productos_por_categoria(datos_tienda)
        resultados[nombre_tienda] = conteo_categorias

        # Calcular el total de ventas en dinero
        total_venta = calcular_total_ventas(datos_tienda)
        totales_ventas[nombre_tienda] = total_venta

        # Calcular el total de ventas por categorías
        ventas_por_categoria = calcular_ventas_por_categorias(datos_tienda)
        ventas_por_categorias[nombre_tienda] = ventas_por_categoria

    # Mostrar los resultados
    mostrar_resultados(resultados, totales_ventas, ventas_por_categorias)

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import pandas as pd

# URLs de los archivos CSV
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv "
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv "
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv "
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv "

# Cargar los datos de cada tienda
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Agregar una columna 'Tienda' a cada DataFrame
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar todos los DataFrames en uno solo
datos_combinados = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Paso 1: Calcular la cantidad de productos vendidos por categoría en cada tienda
ventas_por_categoria = (
    datos_combinados.groupby(['Tienda', 'Categoría del Producto'])
    .size()
    .reset_index(name='Ventas')
)

print("Ventas por categoría en cada tienda:")
print(ventas_por_categoria)

# Paso 2: Identificar las categorías más populares de cada tienda
categorias_mas_populares = (
    ventas_por_categoria.loc[ventas_por_categoria.groupby('Tienda')['Ventas'].idxmax()]
)

print("\nCategorías más populares de cada tienda:")
print(categorias_mas_populares)

# Paso 3: Identificar los 3 artículos más vendidos y los 3 menos vendidos de cada tienda
# Agrupar por Tienda y Producto, y contar el número de ventas
ventas_por_producto = (
    datos_combinados.groupby(['Tienda', 'Producto'])
    .size()
    .reset_index(name='Ventas')
)

# Función para obtener los 3 productos más vendidos y menos vendidos de cada tienda
def top_productos(df, top_n=3):
    mas_vendidos = df.nlargest(top_n, 'Ventas')
    menos_vendidos = df.nsmallest(top_n, 'Ventas')
    return mas_vendidos, menos_vendidos

# Aplicar la función a cada tienda
resultados = {}
for tienda in datos_combinados['Tienda'].unique():
    df_tienda = ventas_por_producto[ventas_por_producto['Tienda'] == tienda]
    mas_vendidos, menos_vendidos = top_productos(df_tienda)
    resultados[tienda] = {'Mas Vendidos': mas_vendidos, 'Menos Vendidos': menos_vendidos}

print("\nProductos más vendidos y menos vendidos por tienda:")
for tienda, datos in resultados.items():
    print(f"\n{tienda}:")
    print("Productos más vendidos:")
    print(datos['Mas Vendidos'])
    print("Productos menos vendidos:")
    print(datos['Menos Vendidos'])

# Paso 4: Identificar los productos que generaron mayores ingresos
# Calcular los ingresos totales por producto (Precio * Ventas)
ingresos_por_producto = (
    datos_combinados.assign(Ingresos=lambda x: x['Precio'] * x['Cantidad de cuotas'])
    .groupby(['Tienda', 'Producto'])['Ingresos']
    .sum()
    .reset_index()
)

# Obtener los productos con mayores ingresos por tienda
productos_mayores_ingresos = {}
for tienda in datos_combinados['Tienda'].unique():
    df_tienda = ingresos_por_producto[ingresos_por_producto['Tienda'] == tienda]
    productos_mayores_ingresos[tienda] = df_tienda.nlargest(3, 'Ingresos')

print("\nProductos que generaron mayores ingresos por tienda:")
for tienda, datos in productos_mayores_ingresos.items():
    print(f"\n{tienda}:")
    print(datos)
```
